```python
import pandas as pd

dataset = pd.read_csv('properties_data.csv')
price_summary = dataset['price'].describe().apply('{:,.2f}'.format)
print(price_summary)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.hist(dataset['price'], bins=15, color='orange')
plt.xlabel('Price')
plt.ylabel('Number of Properties')
plt.title('Distribution of Property Price', fontsize=10, color='black')

plt.show()
```

```python
import pandas as pd

dataset = pd.read_csv('properties_data.csv')
price_summary = dataset['quality'].describe()
print(price_summary)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.hist(dataset['quality'], bins=15,color='darkblue')

# Add the x and y axis titles
plt.xlabel('Quality')
plt.ylabel('Number of Properties')

plt.title('Distribution of Property Quality', fontsize=10, color='black')

plt.show()
```

```python
dataset2 = dataset.drop(['neighborhood', 'longitude', 'latitude','id'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder 
label_encoder = LabelEncoder() 
dataset2['quality'] = label_encoder.fit_transform(dataset2['quality']) 
```

```python
import matplotlib.pyplot as mp 
import pandas as pd 
import seaborn as sns
  
# plotting correlation heatmap 
dataplot = sns.heatmap(dataset2.corr())

# displaying heatmap 
mp.show()
```

```python
import seaborn as sns
 
fig = plt.figure(figsize=(5,10))
ax = sns.heatmap(dataset2.corr()[['price']].abs().sort_values('price', ascending=False), annot=True,annot_kws = {"size":12},cmap='Blues',cbar=False)
ax.set_title('Correlation Coefficient for Price',fontsize=18)
ax.tick_params(axis = "both",labelsize=12)
y_min,y_max = ax.get_ylim()
ax.set_ylim(top=y_max+1)
```
