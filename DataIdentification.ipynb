```python
import pandas as pd

dataset = pd.read_csv('properties_data.csv')
price_summary = dataset['price'].describe().apply('{:,.2f}'.format)
print(price_summary)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.hist(dataset['price'], bins=15, color='orange')
plt.xlabel('Price')
plt.ylabel('Number of Properties')
plt.title('Distribution of Property Price', fontsize=10, color='black')

plt.show()
```

```python
import pandas as pd

dataset = pd.read_csv('properties_data.csv')
price_summary = dataset['quality'].describe()
print(price_summary)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.hist(dataset['quality'], bins=15,color='darkblue')

# Add the x and y axis titles
plt.xlabel('Quality')
plt.ylabel('Number of Properties')

plt.title('Distribution of Property Quality', fontsize=10, color='black')

plt.show()
```
